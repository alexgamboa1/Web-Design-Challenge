```python
#Dependencies 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import CSV
data = pd.read_csv('Resources/cities.csv')
```

```python
#Check DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
plt.scatter(df['Lat'],df['Max Temp'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Max Temperature (F)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Max Temperature')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Max Temperature")
plt.show()
```

```python
plt.scatter(df['Lat'],df['Humidity'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Humidity (%)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Humidity')

plt.tight_layout()


#save fig
plt.savefig("City Latitude vs Humidity")
plt.show()
```

```python
plt.scatter(df['Lat'],df['Cloudiness'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Cloudiness (%)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Cloudiness')
plt.tight_layout()

#save fig
plt.savefig("City latitude vs Cloudiness")
plt.show()
```

```python
plt.scatter(df['Lat'],df['Wind Speed'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Wind Speed (mph)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Wind Speed (11/30/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Wind Speed")
plt.show()
```
